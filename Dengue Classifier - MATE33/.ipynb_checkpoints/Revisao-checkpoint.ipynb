```python
# loading all we need...
import pandas as pd
import numpy as np

from sklearn.metrics import classification_report

from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score

from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.model_selection import GridSearchCV

from sklearn.metrics import f1_score

from sklearn.tree import plot_tree
```

```python
# Dados obtidos a partir do artigo:
# Brazilian Forest Dataset: A new dataset to model local biodiversity
# Ricardo A. Rios, Tatiane N. Rios, Gabriel R. Palma & Rodrigo F. De Mello
# https://www.tandfonline.com/doi/full/10.1080/0952813X.2021.1871972
```

```python
from sklearn.preprocessing import LabelEncoder

dataset = pd.read_csv('dengue_sinan_processed.csv')


# Supondo que a coluna com valores 'ano-mês' se chame 'YEAR_MONTH'
le = LabelEncoder()
dataset['DT_NOTIFIC'] = le.fit_transform(dataset['DT_NOTIFIC'])

dataset['CS_SEXO'] = dataset['CS_SEXO'].replace({'M': 1, 'F': 0, 'I': 2})
dataset['CS_SEXO'] = dataset['CS_SEXO'].replace({'M': 1, 'F': 0, 'I': 2})
dataset['CS_GESTANT'] = dataset['CS_GESTANT'].replace({'N':0,'S':1})

dataset
```

```python
y = dataset['CLASSI_FIN']
dataset = dataset.drop(['CLASSI_FIN'], axis = 1)

X_train,X_test,y_train,y_test = train_test_split(dataset, y, test_size=0.3)
```

```python
### KNN
from sklearn.neighbors import KNeighborsClassifier
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier

# Definindo o intervalo de valores para n_neighbors
param_grid = {'n_neighbors': range(1, 21)}

# Instanciando o modelo KNN
knn = KNeighborsClassifier()

# Executando a busca em grade com validação cruzada
grid_search = GridSearchCV(knn, param_grid, cv=10, scoring='accuracy')
grid_search.fit(X_train, y_train)

# Melhor valor de n_neighbors
best_n_neighbors = grid_search.best_params_['n_neighbors']
best_score = grid_search.best_score_

print(f"Melhor valor para n_neighbors: {best_n_neighbors}")
print(f"Melhor acurácia média durante a validação cruzada: {best_score:.4f}")
```

```python
clf = KNeighborsClassifier(n_neighbors=11)
```

```python
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
y_pred
```

```python
pd.crosstab(y_test, y_pred, rownames=['True'], colnames=['Predicted'], margins=True)
```

```python
### arvore de decisao
from sklearn.tree import DecisionTreeClassifier
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier

# Definindo o intervalo de valores para os hiperparâmetros
param_grid = {
    'criterion': ['gini', 'entropy'],  # Critério para dividir os nós
    'max_depth': [None, 10, 20, 30],    # Profundidade máxima da árvore
    'min_samples_split': [2, 5, 10],    # Número mínimo de amostras necessárias para dividir um nó
    'min_samples_leaf': [1, 2, 4]       # Número mínimo de amostras necessárias em um nó folha
}

# Instanciando o modelo de árvore de decisão
clf = DecisionTreeClassifier()

# Configurando o Grid Search com validação cruzada
grid_search = GridSearchCV(estimator=clf, param_grid=param_grid, cv=10, scoring='accuracy', n_jobs=-1)

# Ajustando o Grid Search aos dados de treino
grid_search.fit(X_train, y_train)

# Melhor combinação de hiperparâmetros
print(f"Melhores hiperparâmetros: {grid_search.best_params_}")
print(f"Melhor acurácia média: {grid_search.best_score_:.4f}")
```

```python
clf = DecisionTreeClassifier(criterion = 'entropy', max_depth = 20, min_samples_leaf = 2, min_samples_split = 10)
```

```python
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
pd.crosstab(y_test, y_pred, rownames=['True'], colnames=['Predicted'], margins=True)
```

```python
f1_score(y_test, y_pred, average='macro')
```

```python
plot_tree(clf)
```

```python
### Random Forest
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV

# Definindo o intervalo de valores para os hiperparâmetros
param_grid = {
    'n_estimators': [50, 100, 150],             # Número de árvores na floresta
    'max_depth': [None, 10, 20, 30],            # Profundidade máxima das árvores
    'min_samples_split': [2, 5, 10],            # Número mínimo de amostras necessárias para dividir um nó
    'min_samples_leaf': [1, 2, 4],              # Número mínimo de amostras necessárias em um nó folha
    'bootstrap': [True, False]                  # Se usar ou não bootstrap amostras
}

# Instanciando o modelo de Random Forest
rf = RandomForestClassifier()

# Configurando o Grid Search com validação cruzada
grid_search = GridSearchCV(estimator=rf, param_grid=param_grid, cv=10, scoring='accuracy', n_jobs=-1)

# Ajustando o Grid Search aos dados de treino
grid_search.fit(X_train, y_train)

# Melhor combinação de hiperparâmetros
print(f"Melhores hiperparâmetros: {grid_search.best_params_}")
print(f"Melhor acurácia média: {grid_search.best_score_:.4f}")
```

```python
clf = RandomForestClassifier()
#Arrumar os hiperparâmetros de acordo com o código de cima
```

```python
y_pred = clf.predict(X_test)
pd.crosstab(y_test, y_pred, rownames=['True'], colnames=['Predicted'], margins=True)
```

```python
clf.fit(X_train,y_train)
```

```python
y_pred = clf.predict(X_test)
pd.crosstab(y_test, y_pred, rownames=['True'], colnames=['Predicted'], margins=True)
```

```python
### XGBOOST
import xgboost as xgb
```

```python
params = {'max_depth':np.arange(1, 5, 1), 
         'eta':np.arange(0.1, 0.5, 0.1)
         }


model = xgb.XGBClassifier(objective="binary:logistic", random_state=42)


cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)
clf = GridSearchCV(model, params, cv = cv, n_jobs = -1, scoring='accuracy')
```

```python
clf.fit(X_train,y_train)
```

```python
clf.best_params_
```

```python
y_pred = clf.predict(X_test)
pd.crosstab(y_test, y_pred, rownames=['True'], colnames=['Predicted'], margins=True)
```
